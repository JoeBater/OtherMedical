```python
def read_section_horizon(lines):
    
    sl = -1
    
    for i in range(len(lines)):
        if lines[i] == 'SECTION_HORIZON':
            break
    
    i+= 1
    #print(i)
    
    while(lines[i] != ''):
        if lines[i][0] == '#':
            pass
        else:
            sl = lines[i]
        i += 1
    return int(sl)

n_days = read_section_horizon(lines)
n_days
```

```python
def read_section_shifts(lines):
    
    sl = -1
    sss = -1
    
    for i in range(len(lines)):
        if lines[i] == 'SECTION_SHIFTS':
            sss = i
            break
    
    cols = lines[sss+1][2:].split(',')
    i+= 2
    #print(i)
    
    df = pd.DataFrame(columns=cols)
    
    while(lines[i] != ''):
        #if lines[i][0] == '#':
        l = list(lines[i].split(','))
        
        dct = {}
        for j in range(len(cols)):
            if j < len(cols)-1:
                dct[cols[j]] = l[j]
            else:
                dct[cols[j]] = list(l[j].split('|'))
            
        #print(dct)
        df = df.append(dct, ignore_index=True)
        i += 1
        
        
    # add an 'off' shift to the shifts dataframe with zero minutes duration
    df = df.append({ 'ShiftID': '-', ' Length in mins': 0 , ' Shifts which cannot follow this shift | separated': []}, 
                   ignore_index=True)
    #print('shifts', df.shape)

    return df, df.values, len(df), list(df['ShiftID'])

shifts, shifts_arr, n_shifts, s_list = read_section_shifts(lines)
shifts
```

```python
def read_section_cover(lines):
    
    sl = -1
    sss = -1
    
    for i in range(len(lines)):
        if lines[i] == 'SECTION_COVER':
            sss = i
            break
    
    cols = lines[sss+1].split(',')
    i+= 2
    #print(i)
    
    df = pd.DataFrame(columns=cols)
    
    dct = {}
    for k in range(i, len(lines)):
        l = list(lines[k].split(','))
        for j in range(len(l)):
            dct[cols[j]] = l[j]
        

        #print(dct)
        df = df.append(dct, ignore_index=True)

    return df, df.values

cover, cover_arr = read_section_cover(lines)
cover
```

```python
create = 'random'

def get_allowed_shifts(n, d):
    
    verbose = 0
    
    max_shift_int = len(shifts_arr)-1
    
    not_allowed = []
    allowed = s_list.copy()
    if '' in allowed:
        allowed.remove('')
        
    max_consecutive = int(staff_arr[n,4])
    min_consecutive = int(staff_arr[n,5])
    count_consecutive = 0
    
    #
    if create == 'allowed':
        
        # shifts followed by
        if d > 0:
            if verbose:
                print('yesterday', s_list[solution[n,(d-1)]])

            # get the shifts that are not allowed given yesterdays assignment
            not_allowed = shifts_arr[solution[n,(d-1)], 2]
            if '' in not_allowed:
                not_allowed.remove('')
            if verbose:
                print('not_allowed', not_allowed)

            for x in not_allowed:
                if x != '':
                    allowed.remove(x)
            if verbose:
                print('allowed', allowed)

            # max consecutive shifts constraints - count number of days working a shift, if >= the amount allowed, then off becomes the only option   
            count_consecutive = 0
            for i in range(d-max_consecutive, d):
                if i >= 0:
                    if solution[n,i] < max_shift_int:
                        count_consecutive += 1
            if count_consecutive >= max_consecutive:
                allowed = ['-']
            if verbose:
                print('count_consecutive', count_consecutive)
                print('max_consecutive', max_consecutive)
                print('allowed', allowed)

            # min consecutive shifts constraints      
            count_consecutive = 0
            for i in range(d-min_consecutive, d):
                if i >= 0:
                    if solution[n,i] < max_shift_int:
                        count_consecutive += 1
            if count_consecutive <= min_consecutive:
                allowed.remove('-')
            if verbose:
                print('count_consecutive', count_consecutive)
                print('min_consecutive', min_consecutive)
                print('allowed', allowed)
        
    
    
    allowed_shifts=[]
    for a in allowed:
        allowed_shifts.append(s_list.index(a))
    
#     print('yesterday', s_list[solution[n,(d-1)]], '\n',
#           'not_allowed', not_allowed, '\n',
#           'allowed', allowed, '\n',
#           'count_consecutive', count_consecutive, '\n',
#           'max_consecutive', max_consecutive, '\n',
#           'min_consecutive', min_consecutive, '\n',
#           'allowed_shifts', allowed_shifts)
    
    return allowed_shifts

x = get_allowed_shifts(9,0)
print(x)
```

```python
def evaluate_staff_shifts():
    
    staff_shifts_cost = 0
    
    for n in range(len(staff)):
        #print('nurse', n)
        shift_counts = [0] * len(s_list)
        for ms in range(len(staff_arr[n, 1])):
            #print(staff.iloc[n, 1][ms])
            #print(staff.iloc[n, 1][ms][0])
            #print(staff.iloc[n, 1][ms][2:])

            shift_counts[s_list.index(staff_arr[n, 1][ms][0])] = int(staff_arr[n, 1][ms][2:])
        #print(shift_counts) 
    
        for d in range(n_days):
            shift_counts[solution[n,d]] -= 1
        
        # we don't care about the 'off' shift
        for ss in range(len(shift_counts)-1):
            if shift_counts[ss] < 0:
                staff_shifts_cost += np.abs(shift_counts[ss]) * int(shifts_arr[ss, 1])
    
    return staff_shifts_cost

evaluate_staff_shifts()
```

```python
def get_nurse_minutes(n):
    
    nurse_minutes = 0
        
    for d in range(n_days):
        nurse_minutes += int(shifts_arr[solution[n,d], 1])
        
    return nurse_minutes

def evaluate_nurse_minutes():
    
    nurse_cost = 0
    
    for n in range(n_nurses):
        
        nurse_minutes = get_nurse_minutes(n)
    
        if nurse_minutes > int(staff_arr[n, 2]):
            nurse_cost += nurse_minutes - int(staff_arr[n, 2])
        if nurse_minutes < int(staff_arr[n, 3]):
            nurse_cost += int(staff_arr[n, 3]) - nurse_minutes
            
        #print(n, nurse_minutes, nurse_cost)
        
    return nurse_cost
        
evaluate_nurse_minutes()
```

```python
def evaluate_shift_cover():

    # evaluate solution
    shift_cover_costs = 0

    # initially use df but shift to array later
    for sc in range(len(cover_arr)):

        shifts_count = 0
        #print(cover_arr[sc, 0])

        for d in range(n_days):
            if d == int(cover_arr[sc, 0]):

                #print(cover_arr[sc, :])

                for n in range(n_nurses):
                    if solution[n,d] == s_list.index(cover_arr[sc, 1]):
                        shifts_count += 1

                #print(d, cover_arr[sc, 0], cover_arr[sc, 1], s_list.index(cover_arr[sc, 1]), int(cover_arr[sc, 2]), shifts_count)

                if shifts_count < int(cover_arr[sc, 2]):
                    shift_cover_costs += (int(cover_arr[sc, 2]) - shifts_count) * int(cover_arr[sc, 3])
                if shifts_count > int(cover_arr[sc, 2]):
                    shift_cover_costs += (shifts_count - int(cover_arr[sc, 2])) * int(cover_arr[sc, 4])

    return shift_cover_costs      
        
evaluate_shift_cover()
```

```python
# run a parallel set of scores for the hard constraints
def evaluate_required_shifts():
    
    disallowed = 0
    
    max_shift_int = len(shifts_arr)-1

    allowed = s_list.copy()
    
    verbose = False
    
    for n in range(n_nurses):
        for d in range(n_days):
            max_consecutive = int(staff_arr[n,4])
            min_consecutive = int(staff_arr[n,5])
            count_consecutive = 0
    
            # shifts followed by
            if d > 0:
                if verbose:
                    print('yesterday', s_list[solution[n,(d-1)]])

                # get the shifts that are not allowed given yesterdays assignment
                not_allowed = shifts_arr[solution[n,(d-1)], 2]
                if '' in not_allowed:
                    not_allowed.remove('')
                if verbose:
                    print('not_allowed', not_allowed)

                for x in not_allowed:
                    if x != '':
                        if x in allowed:
                            allowed.remove(x)
                #if verbose:
                  #  print('allowed', allowed)

                # max consecutive shifts constraints - count number of days working a shift, if >= the amount allowed, then off becomes the only option   
                count_consecutive = 0
                for i in range(d-max_consecutive, d):
                    if i >= 0:
                        if solution[n,i] < max_shift_int:
                            count_consecutive += 1
                if count_consecutive >= max_consecutive:
                    allowed = ['-']
                if verbose:
                    print('count_consecutive', count_consecutive)
                    print('max_consecutive', max_consecutive)
                    print('allowed', allowed)

                # min consecutive shifts constraints      
                count_consecutive = 0
                for i in range(d-min_consecutive, d):
                    if i >= 0:
                        if solution[n,i] < max_shift_int:
                            count_consecutive += 1
                if count_consecutive <= min_consecutive:
                    if '-' in allowed:
                        allowed.remove('-')
                if verbose:
                    print('count_consecutive', count_consecutive)
                    print('min_consecutive', min_consecutive)
                    print('allowed', allowed)    
                    
                allowed_shifts = []
                for a in allowed:
                    allowed_shifts.append(s_list.index(a))
                    
                if solution[n,d] not in allowed_shifts:
                    disallowed += 1
                    
    return disallowed
    
d = evaluate_required_shifts()
d
```

```python
def evaluate_solution():

    return evaluate_shift_cover() + evaluate_nurse_minutes() + evaluate_staff_shifts(), evaluate_required_shifts()

evaluate_solution()
```

```python
def local_search():
    
    best_score = evaluate_solution()
    
    change = 1
    while(change == 1):
        change = 0
        for n in range(n_nurses):
            for d in range(n_days):
                
                allowed_shifts = get_allowed_shifts(n,d)
                #allowed_shifts = s_list[np.random.randint(0, len(s_list))] 
                
                for s in allowed_shifts:
                    
                    if s != solution[n,d]:
                        old_shift = solution[n,d]
                        solution[n,d] = s

                        new_score = evaluate_solution()

                        
                        if new_score[0] < best_score[0] and new_score[1] == best_score[1]:
                            best_score = new_score
                            change = 1
                        elif new_score[0] == best_score[0] and new_score[1] < best_score[1]:
                            best_score = new_score
                            change = 1
                        else:
                            solution[n,d] = old_shift
                        
    #print(best_score)
    

local_search()
evaluate_solution()
```

```python
ITERS = 50
PERTURB = int(n_nurses * n_days * 0.55)

def iterated_local_search(solution):
    
    local_search()
    best_score = evaluate_solution()
    best_solution = solution.copy(order='C')
    
    for iter in trange(ITERS):
        
        for p in range(PERTURB):
            n = np.random.randint(0, n_nurses)
            d = np.random.randint(0, n_days)
            solution[n,d] = np.random.randint(0, n_shifts)
    
        local_search()
        new_score = evaluate_solution()

        if new_score[0] < best_score[0] and new_score[1] == best_score[1]:
            best_score = new_score
            change = 1
        elif new_score[0] >= best_score[0] and new_score[1] < best_score[1]:
            best_score = new_score
            change = 1
        else:
            solution = best_solution.copy()
            
        print(iter, best_score)   
        
iterated_local_search(solution)
```
